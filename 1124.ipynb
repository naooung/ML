### 16.2 원 핫 인코딩
#### 연속 vs 카테고리 값
- 연속값 → 실수로 표현
    - 키, 몸무게 등
- 카테고리 값 → 이산 값. 단어나 클래스로 표현
- 차이점: 비슷한 수의 의마
    - 연속 값: 비슷함을 나타냄
    - 카테고리 값: 일치하지 않음을 나타냄
- 연필-볼펜, 연필-자 사이의 관계(거리)
    - 상식: distance(연필, 볼펜) < distance(연필, 자)
    - 부등호의 방향이 상식과 반대로 될 수 있으므로, 카테고리 값을 특징 벡터로 표현할 때 인덱스를 사용할 수 없다.

#### 원 핫 인코딩
- 한 개의 1과 n-1개의 0으로 구성된 n차원 벡터
- 연속값과 달리 카테고리 값은 특징 벡터로 표시할 때, 벡터간 거리는 의미가 없음
- 원 핫 인코딩은 카테고리를 위해 사용

#### 단어 임베딩
- 자연어 처리에 사용되는 단어 데이터를 표현하기 위한 벡터
- 단어 데이터를 원 핫 인코딩 벡터로 표현하면 단어 간의 유사도를 표현할 수 없음
- Word2Vec 같은 방법을 사용하여 단어 임베딩을 표현함
    - 단어의 의미를 다차원 공간에 벡터화 → 분산 표현
    - 원 핫 인코딩이라는 고차원 벡터에 차원별로 분리해서 표현한 것을 저차원의 고밀도 벡터로 분산해서 표현 

#### 주 성분 분석 (PCA)
- 대표적인 선형 차원 축소 방법
- 데이터를 표현하는 새로운 축을 찾아서 다음 조건이 만족하는 축으로 데이터를 변환
    - 모든 데이터를 새로운 축으로 투사하고, 투사한 데이터들 간의 거리의 합이 최대가 되도록 함
    - 데이터와 새로운 축 간의 직선거리의 총합이 최소가 되도록 함
- 데이터와 새로운 축 간의 직선거리의 총합이 최소가 되도록 함

### 16.3 오토인코더(Autoencoder)
- 비선형 차원 축소 방법
- 인코더와 디코더로 구성
- 인코더는 데이터 압축과 유사하게 입력 데이터는 차원을 축소하여 중간 결과물(z)를 생성함
- 인코더는 입력 샘플의 정보를 최대한 보존하도록 손실 압축을 수행
- 디코더는 압축된 중간 결과물로부터 입력과 최대한 같아지도록 압축을 해제

#### 왜 은닉인가?
- 은닉 벡터는 원하는 출력을 뱉어내는 데 해당 모델만 알고 있는 방법으로 감춰놓고 있다.
- MNIST 샘플 벡터의 첫 번쨰 차원 → 첫 번째 픽셀
- 은닉 벡터의 차원은? → 알 수 없음
