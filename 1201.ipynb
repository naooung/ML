## 18. CNN (합성곱신경망)
- 컴퓨터 버전 문제에 적합한 아키텍쳐

### 18.1 전통적인 방식
- 기존 컴퓨터 비전 방식
    - 미리 정해진 합성곱 커널(필터, 마스크)을 사용하여 원하는 형태의 패턴 추출
- 소벨(sobel) 필터 적용 사례: 경계 검출 → 경계라는 특징 정보 추출

### 18.2 합성곱 연산
- 입력과 같은 크기의 출력을 가지기 위해 패딩을 넣으려면 P = (K - 1) / 2만큼 추가해야 함

#### 패턴 추출의 원리
- 머신러닝 vs 딥러닝

#### 채널
- 입력 채널: 특징을 추출하기 위한 입력이 여러 종류일 때 이것을 입력 채널이라고 함
    - 컬러 이미지의 RGB 채널
- 출력 채널: 여러 가지 특징의 패턴을 동시에 추출하고 싶을 때 여러 개의 커널을 적용하게 되는데, 이 커널의 수를 출력채널이라고 함

#### 합성곱 계층의 특징과 장단점
- 선형 계층
    - 입력 벡터의 각 차원은 별개의 의미를 가진 특징(feature)임
    - 입력 데이터의 자리를 밀려 쓰거나 바뀌거나 하면 데이터의 의미를 잃어버림
        - 키가 들어갈 위치에 몸무게 값을 넣으면 데이터의 의미가 없음
- 합성곱 계층
    - 이미지는 샘플의 픽셀이 약간씩 이동되거나 픽셀의 값이 약간 달라지더라도 인식하는데 문제가 없어야 함
    - 합성곱 계층은 이미지 데이터의 특징에 맞ㄱ게 이런 변화에도 정상젓으로 학습이 가능한 구조임
- 선형 계층
    - 입출력 벡터의 크기가 정해지면 내부 가중치 파라미터의 크기가 정해점
- 합성곱 계층
    - 입출력 벡터의 크기와 상관없이 내부 커널의 크기에 따라 기중치 파라미터의 크기가 정해집
    - 선형 계층보다 훨씬 적은 수의 파라미터를 가장
    - ~ 속도, 에모리 측면에서 이득
    - 최적화 난이도가 낮아짐
- 합성곱 계층의 단점
    - 입출력 크기가 고정되어 있지 않기 때문에 직관적으로 확인하기 어려움
    - 입출력 크기를 계신하는 복잡한 수식 필요
    - 신경망구현 난이도가 높아짐

#### Max Pooling vs 스트라이드
- 스트라이드: 커널을 적용할 때 성큼성큼 칸을 건너뛰어 적용되도록 하는 기법
